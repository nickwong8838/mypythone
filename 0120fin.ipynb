```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# 定义需要分析的企业股票代码
symbols = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'FB']  # 苹果、微软、谷歌、亚马逊、Meta

# 创建空字典存储营业收入和股票价格数据
revenue_data = {}
price_data = {}

# 循环获取每个企业的财务数据和股票价格
for symbol in symbols:
    stock = yf.Ticker(symbol)
    
    # 获取总收入数据
    financials = stock.financials
    if 'Total Revenue' in financials.index:
        revenue = financials.loc['Total Revenue']
        revenue_data[symbol] = revenue
    else:
        print(f"{symbol} 不存在 'Total Revenue' 数据。")
    
    # 获取股票价格数据（使用历史数据）
    price_history = stock.history(period='5y')  # 获取过去5年的数据
    price_data[symbol] = price_history['Close']

# 将数据转换为 DataFrame
revenue_df = pd.DataFrame(revenue_data)
price_df = pd.DataFrame(price_data)

# 检查是否获取到数据
if not revenue_df.empty and not price_df.empty:
    # 计算增长率
    growth_rate_df = revenue_df.pct_change(axis=0) * 100  # 计算百分比变化

    # 打印总收入和增长率数据框
    print("总收入数据框:")
    print(revenue_df)
    print("\n增长率数据框 (%):")
    print(growth_rate_df)

    # 计算相关系数
    correlation_results = {}
    for symbol in symbols:
        if symbol in growth_rate_df.columns and symbol in price_df.columns:
            # 对齐数据并去除缺失值
            combined_df = pd.concat([growth_rate_df[symbol], price_df[symbol]], axis=1)
            combined_df.columns = ['Growth Rate', 'Stock Price']
            combined_df.dropna(inplace=True)  # 去除缺失值
            
            if not combined_df.empty:
                # 计算相关系数
                correlation = combined_df['Growth Rate'].corr(combined_df['Stock Price'])
                correlation_results[symbol] = correlation
            else:
                print(f"{symbol} 的有效数据不足以计算相关系数。")
        else:
            print(f"{symbol} 的数据不足以计算相关系数。")

    # 打印相关系数
    print("\n收入增长率与股价的相关系数:")
    for symbol, correlation in correlation_results.items():
        print(f"{symbol}: {correlation:.3f}")

    # 绘制总收入和股票价格折线图
    plt.figure(figsize=(14, 7))

    # 绘制总收入
    plt.subplot(2, 1, 1)
    for symbol in revenue_df.columns:
        plt.plot(revenue_df.index, revenue_df[symbol], marker='o', label=f'{symbol} 总收入')
    plt.title('总收入折线图')
    plt.xlabel('年度')
    plt.ylabel('总收入 (单位: 亿)')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()

    # 绘制股票价格
    plt.subplot(2, 1, 2)
    for symbol in price_df.columns:
        plt.plot(price_df.index, price_df[symbol], marker='x', label=f'{symbol} 股票价格')
    plt.title('股票价格折线图')
    plt.xlabel('日期')
    plt.ylabel('股票价格 (单位: $)')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()
else:
    print("没有可用的总收入或股票价格数据。")
```
