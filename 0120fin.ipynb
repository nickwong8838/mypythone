```python
import yfinance as yf
import pandas as pd

# 定义需要分析的企业股票代码
symbols = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'FB']  # 苹果、微软、谷歌、亚马逊、Meta

# 创建一个空的字典来存储营业收入数据
revenue_data = {}

# 循环获取每个企业的财务数据
for symbol in symbols:
    stock = yf.Ticker(symbol)
    # 获取总收入数据，按年分组
    revenue = stock.financials.loc['Total Revenue']
    revenue_data[symbol] = revenue

# 将数据转换为DataFrame
revenue_df = pd.DataFrame(revenue_data)

# 计算增长率
growth_rate_df = revenue_df.pct_change(axis=0) * 100  # 计算百分比变化

# 显示结果
print("总收入数据框:")
print(revenue_df)
print("\n增长率数据框 (%):")
print(growth_rate_df)
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# 定义需要分析的企业股票代码
symbols = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'FB']  # 苹果、微软、谷歌、亚马逊、Meta

# 创建一个空的字典来存储营业收入数据
revenue_data = {}

# 循环获取每个企业的财务数据
for symbol in symbols:
    stock = yf.Ticker(symbol)
    financials = stock.financials
    # 检查是否存在 'Total Revenue' 行
    if 'Total Revenue' in financials.index:
        revenue = financials.loc['Total Revenue']
        revenue_data[symbol] = revenue
    else:
        print(f"{symbol} 不存在 'Total Revenue' 数据。")

# 将数据转换为 DataFrame
revenue_df = pd.DataFrame(revenue_data)

# 检查是否获取到数据
if not revenue_df.empty:
    # 计算增长率
    growth_rate_df = revenue_df.pct_change(axis=0) * 100  # 计算百分比变化

    # 绘制总收入折线图
    plt.figure(figsize=(12, 6))
    for symbol in revenue_df.columns:
        plt.plot(revenue_df.index, revenue_df[symbol], marker='o', label=symbol)
    
    plt.title('总收入折线图')
    plt.xlabel('年度')
    plt.ylabel('总收入 (单位: 亿)')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

    # 绘制增长率折线图
    plt.figure(figsize=(12, 6))
    for symbol in growth_rate_df.columns:
        plt.plot(growth_rate_df.index, growth_rate_df[symbol], marker='o', label=symbol)
    
    plt.title('增长率折线图')
    plt.xlabel('年度')
    plt.ylabel('增长率 (%)')
    plt.xticks(rotation=45)
    plt.axhline(0, color='gray', linestyle='--')  # 添加水平线
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()
else:
    print("没有可用的总收入数据。")
```
