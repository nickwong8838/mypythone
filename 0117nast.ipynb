```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# 设置日期范围
period_1 = ("1995-01-01", "2005-12-31")
period_2 = ("2022-01-01", pd.to_datetime("today").strftime("%Y-%m-%d"))

# 获取数据
nasdaq_data_1 = yf.download('^IXIC', start=period_1[0], end=period_1[1])
nasdaq_data_2 = yf.download('^IXIC', start=period_2[0], end=period_2[1])

# 绘制折线图
plt.figure(figsize=(14, 7))

# 1995到2005年数据
plt.subplot(2, 1, 1)
plt.plot(nasdaq_data_1.index, nasdaq_data_1['Close'], label='1995-2005', color='blue')
plt.title('NASDAQ Composite Index (1995-2005)')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid()

# 2022年至今数据
plt.subplot(2, 1, 2)
plt.plot(nasdaq_data_2.index, nasdaq_data_2['Close'], label='2022-Present', color='green')
plt.title('NASDAQ Composite Index (2022-Present)')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
# 计算年化收益率
beginning_value = nasdaq_data_1['Close'].iloc[0]  # 1995年初的收盘价
ending_value = nasdaq_data_1['Close'].iloc[-1]  # 2005年末的收盘价
n_years = (pd.to_datetime(period_1[1]) - pd.to_datetime(period_1[0])).days / 365.25  # 计算年数

annualized_return = (ending_value / beginning_value) ** (1 / n_years) - 1

# 打印年化收益率
print(f"1995到2005年期间的年化收益率为:" ,annualized_return)
```
